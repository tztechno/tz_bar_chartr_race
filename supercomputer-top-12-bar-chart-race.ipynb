# Data preparation

```python
data_dir='../input/top500/top500'
```

```python
files=os.listdir('../input/top500/top500')
print(files)
print(len(files))
```

```python
data = pd.read_csv('../input/top500/top500/June 2021.csv')
data[0:2].T
```

```python
print(data.columns.tolist())
```

```python
cols=['Name']
```

```python
def conj(A):
    A2=A.replace(' - ',' ').replace(', ',' ').replace('  ',' ').split(' ')[0:5]
    a=''
    for item in A2:
        a+=item+' '
    return a
```

```python
data=pd.DataFrame(columns=cols)
for file in files:
    path=os.path.join(data_dir,file)
    datai=pd.read_csv(path)
    datai['file']=file[0:-4]
    datai['Name2']=datai['Name'].apply(lambda x:conj(x))
    data=pd.concat([data,datai])
data=data.reset_index(drop=True)
data
```

```python
print(len(data[data['HPCG [TFlop/s]']=='False']))
print(len(data[data['Linpack Performance (Rmax)']=='False']))
print(len(data[data['Theoretical Peak (Rpeak)']=='False']))
print(len(data[data['Name']=='False']))
```

```python
data['year']=data['file'].apply(lambda x: x.split(' ')[-1])
data2=data[['year','Name2','HPCG [TFlop/s]','Manufacturer:']]
data2=data2.reset_index(drop=True)
data2
```

```python
drop2=data2[data2['HPCG [TFlop/s]']==False].index.tolist()
data3=data2.drop(index=drop2)
drop3=data3[data3['HPCG [TFlop/s]']=='False'].index.tolist()
data4=data3.drop(index=drop3).sort_values('year').reset_index(drop=True)
data4
```

```python
drop=[]
for i in range(1,len(data4)):
    if data4.iloc[i,0]==data4.iloc[i-1,0] and data4.iloc[i,1]==data4.iloc[i-1,1]:
        drop+=[i]
print(drop)
data4=data4.drop(index=drop).reset_index(drop=True)
data4
```

```python
drop=[]
for i in range(len(data4)):
    for j in range(i+1,len(data4)):
        if data4.iloc[i,0]==data4.iloc[j,0] and data4.iloc[i,1]==data4.iloc[j,1]:
            drop+=[j]
print(drop)
data4=data4.drop(index=drop).reset_index(drop=True)
data4
```

```python
data4['HPCG [TFlop/s]']=data4['HPCG [TFlop/s]'].apply(lambda x: float(x.replace(',','')))
```

# Dataframe for Bar Chart Race

```python
# set double indexes and unstack!!!
data4a=data4.set_index(['year','Name2']).unstack()['HPCG [TFlop/s]']
data4a
```

```python
data4a2 = data4a.fillna(0)
data4a2
```

# Create Bar Chart Race


```python
df = data4a2
bcr.bar_chart_race(
    df=df,
    filename='BCR_SupercomputerTop12.mp4',
    orientation='h',
    sort='desc',
    n_bars=12,
    fixed_order=False,
    fixed_max=False,
    steps_per_period=15,
    interpolate_period=False,
    label_bars=True,
    bar_size=.85,
    period_label={'x': .99, 'y': .10, 'ha': 'right', 'va': 'center'},
    #period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,'s': '',
                                      'ha': 'right', 'size': 8, 'family':'IPAGothic'},
    #perpendicular_bar_func='median',
    period_length=1700,
    figsize=(6,4),
    dpi=400,
    cmap='dark24',
    title='Supercomputer Top 12 -- HPCG [TFlop/s]',
    title_size=10,
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family':'IPAGothic','color':'.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha':.8},
    filter_column_colors=True)
```
